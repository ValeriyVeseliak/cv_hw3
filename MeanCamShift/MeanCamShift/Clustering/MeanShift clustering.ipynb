```python
from sklearn.cluster import MeanShift
from matplotlib import pyplot as plt
import matplotlib
import numpy as np
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (20.0, 10.0)
```

```python
def read_file(path):
    file_data = []
    with open(path) as f:
        line = f.readline()
        while line:
            line_arr = line.strip().split()
            result_line = np.array(list(map(lambda x: int(x), line_arr)))
            file_data.append(result_line)
            line = f.readline()
    return np.array(file_data)
```

# S-datasets

Synthetic 2-d data with N=5000 vectors and k=15 Gaussian clusters with different degree of cluster overlap 

### S1 dataset

```python
#s1 
s1_data = read_file("s1.txt")
s1_golden_clusters_centroids = read_file("s1-cb.txt")

s1_clustering = MeanShift(bandwidth=55000).fit(s1_data)
print(len(s1_clustering.cluster_centers_))
print(len(s1_golden_clusters_centroids))
plt.scatter(s1_data[:,0], s1_data[:,1], marker='o', c=s1_clustering.labels_, cmap='Dark2')
```

```python
plt.scatter(s1_clustering.cluster_centers_[:,0], s1_clustering.cluster_centers_[:,1], marker='o', c="blue")
plt.scatter(s1_golden_clusters_centroids[:,0], s1_golden_clusters_centroids[:,1], marker='o', c="red")
```

We can see, that centroids we have found are pretty close to real centroids from our input data clusters. With bandwidth parameter we can adjust density of our clusters and can make our predicted cluster close to real ones.
